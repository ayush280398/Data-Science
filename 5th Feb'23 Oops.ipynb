```python
#Q1
'''In object-oriented programming a class is a blueprint or a template for creating objects. 
It defines the properties and behaviors that objects of that class should have. 
In other words, a class is a user-defined data type that encapsulates data 
and functions related to a specific concept or entity.

An object, on the other hand, is an instance of a class.
It is a tangible entity that can be created based on the defined class. 
Objects have their own unique set of attributes and can perform operations or actions defined by the class's methods. 
In simpler terms an object is a specific realization or manifestation of a class.'''

#Example
class Car:
    def __init__(self, brand,model,year):
        self.brand= brand
        self.model=model
        self.year=year
        
    def accelerate(self):
        print(f"The {self.brand} {self.model} is accelerating.")
        
    def brake(self):
        print(f"The {self.brand} {self.model} is braking.")
     
    def honk(self):
        print(f"The {self.brand} {self.model} is honk")
   
```

```python
#Q2
'''The four pillars of object-oriented programming (OOP) are:

Encapsulation: Encapsulation refers to the bundling of data and the methods that operate on that data into a single unit 
called a class. 
It hides the internal details of an object and provides a well-defined interface to interact with the object. 
Encapsulation helps in achieving data abstraction and data protection by controlling access to the internal state of an object.

Inheritance: Inheritance is the mechanism that allows a class to inherit properties and behaviors from another class.
It enables the creation of a hierarchical relationship between classes 
where a derived class (child class) can inherit and extend the features of a base class.
Inheritance promotes code reuse, modularity, and the concept of "is-a" relationship.

Polymorphism: Polymorphism allows objects of different classes to be treated as objects of a common parent class.
It provides a way to perform a single action in different ways based on the object type or class.
Polymorphism enables code to be written that can work with objects of different types, providing flexibility and extensibility 
to the program.

Abstraction: Abstraction involves the concept of representing complex real-world entities in a simplified manner in software 
systems. It focuses on capturing only the essential characteristics of an object and hiding unnecessary details.
Abstraction allows programmers to create abstract classes and interfaces that define a set of common behaviors 
without providing implementation details. It helps in managing complexity, promoting code reusability, and 
supporting modular design. '''
```

```python
#Q3

'''The __init__() function, also known as the constructor, is a special method in Python classes.
It is automatically called when an object of the class is created. The primary purpose of the __init__() function is to 
initialize the attributes (data members) of an object with the values provided during object creation.'''
#Example

class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def introduce(self):
        print(f"Hello, my name is {self.name} and I'm {self.age} year old.")
        
Person1=Person("john",25) 

Person1.introduce()
```

```python
#Q5
'''Inheritance is a fundamental concept in object-oriented programming that allows a class to inherit properties and behaviors 
from another class. 
The class that inherits from another class is called the derived class or child class, 
and the class being inherited from is called the base class or parent class. 
Inheritance promotes code reuse, modularity, and the concept of an "is-a" relationship between classes.'''

''' Single Inheritance:
Single inheritance refers to the case where a class inherits from a single base class. 
The derived class inherits all the attributes and methods of the base class.'''
#ex (Single Inheritance)
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def drive(self):
        print("Driving the vehicle.")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def accelerate(self):
        print("Accelerating the car.")


car = Car("Toyota", "Camry")

print(car.brand)    
car.drive()        
car.accelerate()   

#Ex(Multiple Inheritance)
''' Multiple inheritance allows a class to inherit from multiple base classes.
The derived class inherits attributes and methods from all the base classes.'''

class Animal:
    def breathe(self):
        print("Animal is breathing.")

class Mammal:
    def feed_milk(self):
        print("Mammal is feeding milk.")

class Dolphin(Animal, Mammal):
    def swim(self):
        print("Dolphin is swimming.")
dolphin = Dolphin()

dolphin.breathe()      
dolphin.feed_milk()    
dolphin.swim()

#Ex(Multilevel Inheritance)
''' Multilevel inheritance occurs when a derived class inherits from another derived class.
The derived class inherits attributes and methods from its immediate base class 
which in turn inherits from its own base class, and so on.'''

class Animal:
    def breathe(self):
        print("Animal is breathing.")

class Mammal(Animal):
    def feed_milk(self):
        print("Mammal is feeding milk.")

class Dolphin(Mammal):
    def swim(self):
        print("Dolphin is swimming.")
dolphin = Dolphin()

dolphin.breathe()      
dolphin.feed_milk()    
dolphin.swim()          
```
